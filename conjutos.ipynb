```python
# a)
A = {0, 1, 2, 3}
B = {2, 3, 4, 5}
print("a)", A.union(B))  # A ∪ B

# b)
A = {-1, 0, 2, 4, 5}
B = {-2, -1, 1, 3, 4, 6}
print("b) A - B:", A.difference(B))
print("b) B - A:", B.difference(A))

# c)
A = {0, 1, 2, 3, 4, 5, 6}
B = {1, 2}
C = {2, 3, 4}
D = {4, 5}
resultado_c = A.difference(B).intersection(C.intersection(D))
print("c)", resultado_c)

# d)
A = {0, -1, 1}
B = {1, 3, 4}
C = {0, 1}
resultado_d = A.difference(B).intersection(C.difference(B))
print("d)", resultado_d)

# e)
A = {-1, 0, 1}
B = {-2, 0, 1, 2}
C = {1, 2, 3, 4, 5}
E = A.union(B).intersection(B.union(C))
print("e)", E)

# f)
A = {-2, -1, 0, 1, 2}
B = {0, 2, 4, 6}
C = {0, 1, 2, 3, 4, 5}
resultado_f = C.difference(A.intersection(B))
print("f)", resultado_f)

# g)
A = {1, 2, 3}
B = {2, 3, 4, 5}
C = {0, 1, 2, 3, 4}
parte1 = A.union(B).intersection(C)
parte2 = A.intersection(B).union(C)
print("g)", parte1.difference(parte2))

# h)
A = {1, 2, 3}
B = {2, 3, 4}
C = {4, 5}
X = A.intersection(B).union(C.difference(A.union(B)))
print("h)", X)

# i)
M = {1, 2, 3, 4, 5}
N = {1, 2, 3}
print("i) M ∪ N:", M.union(N))
print("i) M ∩ N:", M.intersection(N))

# j)
A = {'a', 'b', 'c', 'd', 'e', 'f', 'g'}
B = {'b', 'd', 'g', 'h', 'i'}
C = {'e', 'f', 'm', 'n'}
resultado_j = A.difference(B).intersection(B.difference(C))
print("j)", resultado_j)
```

```python
A = {0, 1, 2, 3}
B = {-2, -1, 0, 1, 2}
C = {-1, 0, 1, 2, 3, 4, 5}

print("a)", B.union(A) == C)                    # A ∪ B == C
print("b)", A.intersection(B) == {0, 1, 2})     # A ∩ B == {0, 1, 2}
print("c)", B.difference(C) == {3, 4, 5})       # B - C == {3, 4, 5}?

print("d)", A.union(C) == C)                    # A ∪ C == C
print("e)", B.intersection(C) == {-1, 0, 1, 2}) # B ∩ C == {-1, 0, 1, 2}
print("f)", A.intersection(B).intersection(C) == {0, 1, 2})  # A ∩ B ∩ C == {0, 1, 2}

print("g)", A.union(B).difference(C) == set()) # (A ∪ B) - C == ∅ ?
print("h)", A.difference(C) == set())          # A - C == ∅ ?
print("i)", B.difference(C.union(A)) == {-2, -1, 4, 5})  # B - (C ∪ A) == {-2, -1, 4, 5}?
print("j)", A.intersection(C.difference(B)) == {3})      # A ∩ (C - B) == {3} ?
```
